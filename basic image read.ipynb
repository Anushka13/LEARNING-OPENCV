```python
import cv2
```

```python
pupimg=cv2.imread("C:/Users/anushka.chitransh/Downloads/puppy.jpeg",0)
print(pupimg)
cv2.imshow('Image1',pupimg)
k=cv2.waitKey(0) & 0xFF
#argument for waitKey is in millisec. If argument is 0, then window will not close on its own. Use cv2.destroyWindow()
#or cv2.destroyAllWindows to destroy the image window
if k==27:
    #27 for escape character
    cv2.destroyAllWindows()
elif k==ord('s'):
    cv2.imwrite("pupimg2.jpeg",pupimg)
    cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0) #0 is the device index, video name can also be written here for an already saved one
print(cap.isOpened())#gives true if device index or video name and location is correct
fourcc=cv2.VideoWriter_fourcc(*'DIVX')
out=cv2.VideoWriter('firstvid.avi',fourcc,20,(640,480))
while(True):
    ret,frame=cap.read()
    if ret==True:
        grayframe=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('hi',grayframe)
        out.write(frame)
        #print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        #print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
    
```
